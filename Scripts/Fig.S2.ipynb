```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib as mpl

# Set up the styling for publication-quality figures - Nature style
plt.rcParams.update({
    'font.size': 10,
    'axes.linewidth': 1.0,
    'axes.labelsize': 16,
    'axes.titlesize': 18,
    'xtick.labelsize': 14,
    'ytick.labelsize': 14,
    'legend.fontsize': 12,
    'figure.dpi': 300,
    'savefig.dpi': 600,
    'savefig.bbox': 'tight',
    'savefig.pad_inches': 0.05,
})

#  
samples = ['functional\ngroups', 'randomly mixed\ngroups']  #    
energy_mev = [1.6581806922287128, 3.3847686489957787]
force = [106.93323986835361, 141.04465200334553]

# Nature     
energy_color = '#95c8d8'  #  
force_color = '#f4a582'  #  /

#   
fig, ax = plt.subplots(figsize=(7, 6))

#     
bar_width = 0.35
x = np.arange(len(samples))

#      
rects1 = ax.bar(x - bar_width/2, energy_mev, bar_width, color=energy_color, edgecolor='black', 
         linewidth=0.8, label='Energy (meV)')

#     y 
ax2 = ax.twinx()
rects2 = ax2.bar(x + bar_width/2, force, bar_width, color=force_color, edgecolor='black', 
         linewidth=0.8, label='Force (meV/)')

#   
ax.set_ylabel('Energy (meV)', color='black')
ax2.set_ylabel('Force (meV/)', color='black')

# x  
ax.set_xticks(x)
ax.set_xticklabels(samples, rotation=0, ha='center')

#  
ax.grid(False)

#  
# lines1, labels1 = ax.get_legend_handles_labels()
# lines2, labels2 = ax2.get_legend_handles_labels()
# ax.legend(lines1 + lines2, labels1 + labels2, loc='upper center', frameon=True, 
#           framealpha=0.7, fontsize=12)

# y  
ax.tick_params(axis='y', colors='black')
ax2.tick_params(axis='y', colors='black')

#   - Nature     
ax.set_facecolor('white')
fig.patch.set_facecolor('white')

# y=0  
ax.axhline(y=0, color='black', linestyle='-', linewidth=0.8, alpha=0.5)

#    -   
for a in [ax, ax2]:
    a.spines['top'].set_visible(True)
    a.spines['right'].set_visible(True)
    a.spines['left'].set_linewidth(1.0)
    a.spines['right'].set_linewidth(1.0)
    a.spines['top'].set_linewidth(1.0)
    a.spines['bottom'].set_linewidth(1.0)

#   
ax.spines['left'].set_color(energy_color)
ax.spines['right'].set_color(force_color)

#    -    
ax.set_xlim(-0.5, len(samples) - 0.5)

#   
plt.subplots_adjust(bottom=0.2)

#   
plt.tight_layout()
plt.savefig('histogram_combined_nature_style.png', dpi=600)
# plt.savefig('histogram_combined_nature_style.pdf', format='pdf')
plt.show()
```
